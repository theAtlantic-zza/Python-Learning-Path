```python
import pandas as pd
"""构建DataFrame的方式"""
# 方式1：列表里放字典
# 特点：
# 1. 最常用
# 2. 一行就是一条数据
# 3. 适合“学生成绩/订单/用户信息”这种表格数据
df1 = pd.DataFrame([
    {"姓名": "小王", "语文": 90, "数学": 95, "英语": 88},
    {"姓名": "小李", "语文": 84, "数学": 91, "英语": 86},
    {"姓名": "小张", "语文": 92, "数学": 89, "英语": 93}
])

print("方式1：列表套字典")
print(df1)
print()

# 方式2：字典里放列表
# 特点：
# 1. 一列就是一个列表
# 2. 更适合“我已经分别拿到了每一列数据”
# 3. 所有列表长度必须一致

df2 = pd.DataFrame({
    "姓名": ["小王", "小李", "小张"],
    "语文": [90, 84, 92],
    "数学": [95, 91, 89],
    "英语": [88, 86, 93]
})

print("方式2：字典套列表")
print(df2)
print()


# 方式3：二维列表 + columns
# 特点：
# 1. 数据本体是纯二维数组
# 2. 需要手动指定列名
# 3. 适合“数据本来就长得像表格矩阵”
df3 = pd.DataFrame(
    [
        ["小王", 90, 95, 88],
        ["小李", 84, 91, 86],
        ["小张", 92, 89, 93]
    ],
    columns=["姓名", "语文", "数学", "英语"]
)

print("方式3：二维列表 + columns")
print(df3)
print()

# 方式4：指定 index
# 特点：
# 1. 可以自己定义行索引
# 2. 适合“每一行有唯一编号/标签”的场景
# 3. 比如学号、订单号、日期
df4 = pd.DataFrame(
    {
        "姓名": ["小王", "小李", "小张"],
        "语文": [90, 84, 92],
        "数学": [95, 91, 89],
        "英语": [88, 86, 93]
    },
    index=["学生1", "学生2", "学生3"]
)

print("方式4：带自定义索引")
print(df4)
print()

# 方式5：先建 Series，再合成 DataFrame
# 特点：
# 1. 适合“先分别处理每一列，再拼起来”
# 2. Series 可以理解成带索引的一列数据
# 3. 更适合稍复杂的数据处理
name = pd.Series(["小王", "小李", "小张"])
chinese = pd.Series([90, 84, 92])
math = pd.Series([95, 91, 89])
english = pd.Series([88, 86, 93])

df5 = pd.DataFrame({
    "姓名": name,
    "语文": chinese,
    "数学": math,
    "英语": english
})
print("方式5：Series 组合成 DataFrame")
print(df5)
print()

# 总结：
# 列表套字典：按“行”构建，最直观
# 字典套列表：按“列”构建，适合已有列数据
# 二维列表 + columns：适合纯表格矩阵数据
# 指定 index：适合有行标签的数据
# Series 组合：适合先处理列，再拼表
```

```python
'''Data Frame的属性 attribute'''
# df.index
# 表示“行索引”,也就是每一行的编号
# 默认是 0, 1, 2, 3 ...
# 也可以自己指定成学号、日期、姓名等
print(df4.index)
print(df4.index.tolist())

# df.columns
# 表示“列名”,也就是表格最上面那一排字段名
# 比如：姓名、语文、数学、英语
print(df4.columns.tolist())

# df.dtypes
# 表示“每一列的数据类型”,注意是 dtypes，不是 dtyoes
# 它会告诉你每一列是整数、浮点数、字符串等类型
print(df4.dtypes.tolist())

# df.size
# 表示“表格里中数据个数”
# 计算方式 = 行数 × 列数
# 比如 3 行 4 列，size 就是 12
print(df2.size)

# df.shape
# 表示“表格的形状”,结果是一个元组：(行数, 列数)
# 比如 (3, 4) 就表示 3 行 4 列
print(df4.shape)
df4.values.tolist()

#df.columns
#获取列名,data frame 特有属性
print(df4.columns.tolist())
```

```python
"""构建series的方式"""
# 方式1：用列表构建 Series
# 最常用,默认索引是 0, 1, 2 ...
s1 = pd.Series([90, 85, 88])

# 方式2：用列表构建 Series，并手动指定索引
# 适合给每个数据配上名字/标签
s2 = pd.Series([90, 85, 88], index=["小王", "小李", "小张"])

# 方式3：用字典构建 Series
# 字典的键会变成索引，值会变成数据
s3 = pd.Series({
    "小王": 90,
    "小李": 85,
    "小张": 88
})

# 方式4：指定统一的值
# 适合快速生成一列相同数据
s4 = pd.Series(100, index=["语文", "数学", "英语"])

print("s1:")
print(s1)
print()

print("s2:")
print(s2)
print()

print("s3:")
print(s3)
print()

print("s4:")
print(s4)
print()

# 总结：
# 列表构建：最基础，默认索引
# 列表+index：自己指定索引
# 字典构建：键做索引，值做数据
# 单个值+index：生成一列相同的数据
```

```python
"series的属性"
print(s3.index.tolist())#获取索引   tolist:把pandas里的数据转成列表
print(s3.values.tolist())#获取值
print(s3.size)#获取数据类型
print(s3.dtype)#获取数据个数
print(s3.shape)#获取数据维度(行,列)
```

```python
from psutil import users
"""数据读取和写入   基于pandas中提供的api,可以很方便的对各类数据文件(csv,excel,数据库,网络数据等)进行读取和写入"""
import pandas as pd
#读取数据
df = pd.read_csv('data/sales.csv',usecols=['订单号','产品类别','产品名称','销售数量','单价'])
#赋值
df['销售金额']=df['销售数量'] * df['单价']#添加一列 并为其赋值
#写入数据 --> to_csv
df.to_csv('data/sales_01.csv',index=False) #索引列不写入
df
# read_csv()：读取 csv 文件
# to_csv()：将处理后的数据写入新的 csv 文件
# index=False：写出时不保存索引列
```

```python
'''数据查看,选择及过滤'''

# 查看
print(df.head(2))   #df.head(n) 查看前n行数据
print(df.tail(2))   #df.tail(n) 查看结尾n行数据
print(df.describe())#df.describe()数值列的统计描述
print(df.info())    #df.info() 查看数据信息(列名,非空计数,数据类型)
print(df.shape)     #df.shape  属性,查看数据维度
print(df.columns)   #df.columns
```

```python
#数据选择
df1 = pd.read_csv('data/sales.csv',usecols=['订单号','产品类别','产品名称','销售数量','单价'])

#1.选择列 纵向
#1.1单列
print(df1['产品名称'].head(2))#df.列名 操作单列
#1.2多列
print(df1[['产品名称', '单价']])#操作多列

#2.选择行 -iloc,loc 横向 切片
#2.1 iloc ---》基于行号选择行(不包含结束位置) !=索引 索引可以人为指定
print(df.iloc[0:5:1] )  # df.iloc[start:stop:step] 按位置切片，不含stop
#2.2 loc  ---》基于行索引选择行(包含结束位置)  index受人为指定影响
df.loc[0:5:1]           # df.loc[start:stop:step]  按索引切片，含stop
```

```python
"""过滤---按条件筛选数据"""
# 条件筛选常用写法
# 比大小 >  <  >=  <=  ==  !=
# isin([...])：判断值是否在指定列表中
# between(a, b)：判断值是否在 a 到 b 之间
# 多条件：& 表示且，| 表示或
# 注意：多个条件时，每个条件都要加括号

# 大于等于：筛选销售数量大于等于 5 的数据
print(df[df['销售数量'] >= 5].head(2))

# isin()：筛选产品类别在“食品”和“图书”中的数据
print(df[df['产品类别'].isin(['食品', '图书'])].head(2))

# between()：筛选销售数量在 3 到 8 之间的数据
print(df[df['销售数量'].between(3, 8)].head(2))

# 且：筛选产品类别是“食品”且销售数量大于等于 3 的数据
print(df[(df['产品类别'] == '食品') & (df['销售数量'] >= 3)].head(2))

# 或：筛选产品类别是“食品”或“图书”的数据
print(df[(df['产品类别'] == '食品') | (df['产品类别'] == '图书')].head(2))
```
